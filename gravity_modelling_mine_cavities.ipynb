```python
import numpy as np  # outils numériques
import matplotlib.pyplot as plt  # graphiques
```

```python
G = 6.674e-11  # m^3 kg^−1 s^−2
x = np.arange(-100, 100, 1)  # position des stations de mesure allant de -100 à 100 m
z = 0  # élévation du gravimètre en m (en surface)
```

```python
# Fonction donnant l'équation du cylindre
def cylindre(x, z, x0, z0, R, delta_rho):
    return 2*np.pi*G*R**2*delta_rho / ( (z - z0)*(1 + ( (x-x0) / (z-z0) )**2) )
```

```python
# Tunnel rempli d’eau

# Paramètre
R_eau = 5  # m 
rho_eau = -1600  # kg/m^3
x0_eau = -25  # m 
z0_eau = -15  # m

# Réponse gravitmétrique
gz_eau = cylindre(x, z, x0_eau, z0_eau, R_eau, rho_eau)

plt.plot(x, gz_eau, lw=2)
plt.xlabel('$x$ (m)')
plt.ylabel('$\Delta g_z$ (m/s$^2$)')
plt.show()
```

```python
# Tunnel rempli d'air

R_air = 5  # m 
rho_air = -2600  # kg/m^3
x0_air = +25  # m 
z0_air = -20  # m

gz_air = cylindre(x, z, x0_air, z0_air, R_air, rho_air)

plt.plot(x, gz_eau, lw=2, label="Eau")
plt.plot(x, gz_air, lw=2, label="Air")
plt.xlabel('$x$ (m)')
plt.ylabel('$\Delta g_z$ (m/s$^2$)')
plt.legend()
plt.show()
```
